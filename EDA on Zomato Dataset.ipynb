```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country
```

```python
final_df=pd.merge(df,df_country,on="Country Code",how="left")
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_vals=final_df.Country.value_counts().values
```

```python
print(country_names)
country_vals
```

```python
plt.pie(country_vals,labels=country_names)
```

```python
#top 3 country that uses zomato
plt.pie(country_vals[:3],labels=country_names[:3],autopct="%1.2f%%")
```

Top 3 country that uses zomato is India, USA, United Kingdoms

```python
##creating rating dataframe
rating=final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating count"})
```

```python
rating
```

## Observation
1. Most people did not the app
2. Excellent ---> 4.5 to 4.9
3. Very Good ---> 4.0 to 3.4
4. Good      ---> 3.5 to 3.9
5. Average   ---> 2.5 to 3.4
6. Poor      ---> 1.4 to 2.4
7. Not Rated ---> 0

```python

sns.barplot(x="Aggregate rating",y="Rating count",hue="Rating color",data=rating,palette=["blue","red","orange","yellow","green","green"])
```

## Observation
1. Not Rated count is very high
2. Maximum number of Ratin are between 2.8 to 3.9

```python
sns.countplot(x="Rating color",hue="Rating color",data=rating,palette=["blue","red","orange","yellow","green","green"])
```
